```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup
import pprint
import json
```

```python
# Set up browser
browser = Browser('chrome')
```

```python
# Visit the Mars news site
url = 'https://coinmarketcap.com/currencies/bitcoin/historical-data'
browser.visit(url)
```

```python
# Function to dismiss overlay (e.g., cookie banner)
def dismiss_overlay():
    try:
        # Look for the overlay dismiss button (adjust selector if needed)
        overlay_button = browser.find_by_css('#onetrust-accept-btn-handler')  # Example for cookie banners
        if overlay_button:
            overlay_button.click()
            print("Overlay dismissed.")
    except Exception as e:
        print(f"No overlay to dismiss: {e}")

# Main loop to click "Load More" button
for i in range(500):  # Adjust the range as needed
    try:
        dismiss_overlay()  # Dismiss any overlay before proceeding
        
        html = browser.html
        soup = BeautifulSoup(html, 'html.parser')
        
        #browser.links.find_by_partial_text('Load').click()
        # Find the "Load More" button by its class or attributes
        
        button = soup.find('button', class_='sc-7d96a92a-0 hOXHRi', string='Load More')
        button = browser.find_by_text('Load More')
        # Click the button if it exists
        if button:
            browser.find_by_css('button.sc-7d96a92a-0.hOXHRi').click()
        else:
            print('Button not found')
    except Exception as e:
        print(f'Error encountered: {e}')


#<button variant="default" data-sensors-click="true" class="sc-7d96a92a-0 hOXHRi" style="padding: 12px 95px;">Load More</button>
```
