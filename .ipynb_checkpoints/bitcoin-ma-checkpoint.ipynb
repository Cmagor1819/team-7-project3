```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# load data
data = pd.read_csv(r'Resources/cleaned_data.csv')
```

```python
# Get the last / recent 30 rows
last_30_days = data.head(30)
last_30_days
```

```python
plt.figure(figsize=(10, 6))
plt.plot(last_30_days['t﻿imeHigh'], last_30_days['timeLow'], marker='o', linestyle='-', color='b')

# Format the x-axis for better readability
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=5))
plt.xticks(rotation=45)

# Add titles and labels
plt.title('Bitcoin Price Change Over Last 30 Days')
plt.xlabel('Date')
plt.ylabel('Price in USD')
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['timeClose'], data['close'], marker='o', linestyle='-', color='blue')

# Format the x-axis for better readability (dates will be spaced every 5 days)

plt.xticks(rotation=45)

# Add titles and labels
plt.title('Bitcoin Closing Prices Over the Last 30 Days', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price in USD', fontsize=12)

# Add grid for easier readability
plt.grid(True)

# Ensure tight layout to prevent cutting off labels
plt.tight_layout()

# Show the plot
plt.show()
```
